```python
import pandas as pd
import numpy as np
import plotly as py
import plotly.graph_objs as go
py.offline.init_notebook_mode(connected=True)
from scipy.signal import argrelextrema
```

```python
nats_season=pd.read_csv('Before_Bryce_Harper_Decision.csv')
```

```python
# nats_season[nats_season['median_ticket_price']==nats_season['median_ticket_price'].min()]
```

```python

random_x = nats_season['date']
random_y = nats_season['median_ticket_price']

trace = go.Scatter(
    x = random_x,
    y = random_y
)
layout = {
    'shapes': [
        # Line Vertical
        {
            'type': 'line',
            'x0': '2019-03-28',
            'x1':'2019-03-28',
            'y0': 0,
            'y1':250,
            'line': {
                'color': 'green',
                'width': 3,
                'dash':'dashdot'
            },
        },
       
    ]
}

data = [trace]
fig = {
    'data': data,
    'layout': layout,
}
py.offline.iplot(fig)
```

```python
local_maxima=argrelextrema(nats_season['median_ticket_price'].values, np.greater)
nats_season.iloc[local_maxima]
```

```python
reg_season=nats_season[nats_season['date']>='2019-03-28']
```

```python
avg_price_by_opp=pd.DataFrame(reg_season.groupby(['opponent'])['median_ticket_price'].mean())
```

```python
avg_price_by_opp
```

```python
avg_price_by_opp.reset_index(inplace=True)
```

```python
formatted_price=[]
for i in range(len(avg_price_by_opp['median_ticket_price'])):
    formatted_price.append('${:,.2f}'.format(avg_price_by_opp['median_ticket_price'][i]))
avg_price_by_opp['formatted_median_ticket_price']=formatted_price
```

```python
# avg_price_by_opp['formatted_median_ticket_price']
```

```python
data = [go.Bar(
            x=avg_price_by_opp['opponent'],
            y=avg_price_by_opp['median_ticket_price'],
            text=(avg_price_by_opp['formatted_median_ticket_price']),
            textposition = 'auto',
            marker=dict(
                color='rgb(158,202,225)',
                line=dict(
                    color='rgb(8,48,107)',
                    width=1.5),
            ))]

py.offline.iplot(data)
```
